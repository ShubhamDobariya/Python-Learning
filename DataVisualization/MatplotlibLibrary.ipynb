```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Matplotlib Basic

```python
x = np.linspace(0,5,11)
x
```

```python
y = x ** 3
y
```

```python
plt.plot(x, y)
plt.title("Hello")
plt.xlabel("X axis")
plt.ylabel("Y axis")
```

### Subplots in Matplotlib

```python
plt.subplot(2,2,1)
plt.plot(x,y**2)
plt.subplot(2,2,4)
plt.plot(x,y)
plt.subplot(2,2,3)
plt.plot(y,x)
plt.subplot(2,2,2)
plt.plot(x**3,y)
```

### Object Oriented Way

```python
fig = plt.figure()
axis1 = fig.add_axes([0.1,0.1,1,1])
axis1.plot(x,y)
axis2 = fig.add_axes([0.3,0.5,0.4,0.4])
axis2.plot(x,y)
```

```python
fig = plt.figure(figsize=(10,8))
axis1 = fig.add_axes([0.1,0.1,1,1])
axis1.plot(x,y)
axis2 = fig.add_axes([0.3,0.5,0.4,0.4])
axis2.plot(x,y)
```

### Types of Plots

### Saving Plots

```python
fig = plt.figure()
ax = fig.add_axes([0.1,0.1,0.8,0.8])
ax.plot(x,y)
plt.savefig("basicplot.png")
```
