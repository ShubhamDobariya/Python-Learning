Read Json to CSV

```python
import pandas as pd
from io import StringIO
```

```python
Data = '{"employee_name":"james", "email":"james@gmail.com", "job_profile":[{"title1":"Team Lead", "title2":"Deleloper"}]}'
```

```python
Json_data = StringIO(Data)
```

```python
pd.read_json(Json_data)
```
