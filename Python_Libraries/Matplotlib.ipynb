MatplotLib Tutorial
- matplotlib is a plotting library for the Python programming language and its numerical mathematics extension Numpy.it provide an object-oriented API for embedding plot into applications using general-purpose GUI toolkit like Tkinter, wxPython, Qt or GTK+. 

```python
import matplotlib.pyplot as plt

# if we use this perticular line then don't use "plt.show()" line
%matplotlib inline   
```

```python
import numpy as np
```

```python
x = np.arange(0,10)
y = np.arange(11,21)
```

```python
a = np.arange(40,50)
b = np.arange(50,60)
```

```python
## plotting using matplotlib

## plt scatter

plt.scatter(x,y, c='g')
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("2D Scatter Graph")
plt.savefig("test.png")
```

```python
y=x*x
```

```python
## plt plot

plt.plot(x,y,"go",linestyle="dashed",linewidth=2,markersize=10)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("2D plot Graph")
```

```python
## Create subplots

plt.subplot(2,2,1)
plt.plot(x,y,"r*--")
plt.subplot(2,2,2)
plt.plot(x,y,"go--")
plt.subplot(2,2,3)
plt.plot(x,y,"b--")
plt.subplot(2,2,4)
plt.plot(x,y,"y")
```

```python
x = np.arange(1,11)
y = 3 * x + 5

plt.title("Matplotlib Demo")
plt.xlabel("X axis caption")
plt.ylabel("Y axis caption")
plt.plot(x,y,"g")
```

```python
## compute the x and y coordinates for points on a sine curve

x = np.arange(0, 4*np.pi, 0.1)
y = np.sin(x)

plt.plot(x,y)
```

```python
## subplot
## compute the x and y coordinates for points on a sine and cosine curves

x = np.arange(0, 5*np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

## set up a subplot grid that has height 3  and width 1
plt.subplot(3,1,1)
plt.plot(x,y_sin,"b")
plt.title("sine")

plt.subplot(3,1,3)
plt.plot(x,y_cos,"g")
plt.title("cosine")

```

```python
## Bar plot 

x = [2,8,10]
y = [11,16,9]

x1 = [3,9,11]
y1 = [6,15,7]

plt.bar(x,y)
plt.bar(x1,y1, color="g")
plt.xlabel("X axis")
plt.ylabel("Y axis")
```

Histograms

```python
x = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27])

plt.hist(x)
plt.title("Histogram")
```

Box Plot using Matplotlib

```python
data = [np.random.normal(0, std, 100) for std in range(1,4)]

plt.boxplot(data, vert=True, patch_artist=True)
plt.title("BoxPlot")
```

Pia Chart

```python
## data to plot

labels = "Python","C++","Ruby","Java"
sizes = [215,130,245,210]
colors = ["gold","yellowgreen","lightcoral","lightskyblue"]
explode = (0.1,0,0,0)

plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True)
plt.axis('equal')
```
