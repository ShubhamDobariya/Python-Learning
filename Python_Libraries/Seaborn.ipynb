## Correlation with Heatmap

- A correlation heatmap uses colored cells, typically in a monochomatic scale, to show a 2D correlation matrix (table) between two discrete or event types. It is very important in Feature Selection

### JoinPlot

- A join plot allows to study the relationship between 2 numeric variable. The central chart display their correlation. It is usually a scatterplot, a hexbin plot, a 2D histogram or a 2D density plot

Univariate Analysis

```python
sns.jointplot(x='tip', y='total_bill', data=df, kind='hex')
```

```python
sns.jointplot(x='tip', y='total_bill', data=df, kind='reg')
```

### Pari plot 

- A "pair plot" is also known as a scatterplot. In which one variable in the same data row is matched with another variable's value, like this: Pairs plot are just elaborations on this, showing all variables paired with all the other variables

```python
sns.pairplot(df)
```

```python
sns.pairplot(df, hue='sex')
```

```python
df['smoker'].value_counts()
```

### Dist Plot

- Dist Plot helps us to checks the distribution of the columns feature

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'], kde=False, bins=10)
```

# Categorical Plot
Seaborn also helps us in doing the analysis on **categorical data point**. In this section we will discuss about
* boxplot
* violinplot
* countplot
* bar plot

```python
## count plot

sns.countplot(x='day',hue='day', data=df,palette="bright")
```

```python
sns.countplot(y='smoker',hue='smoker', data=df, color='g', width=0.5,palette="bright")
```

```python
## Bar plot

sns.barplot(y='total_bill',x='smoker',hue='smoker',data=df,palette="bright")
```

### Box Plot

A box and whisker plot (sometimes called a boxplot) is a graph that presents information from a five-number summary.

```python
sns.boxplot(x='sex',y='total_bill', hue='sex', data=df,palette="bright")
```

```python
sns.boxplot(x='day',y='total_bill', data=df, palette='rainbow')
```

```python
sns.boxplot(data=df,orient='v',palette='rainbow')
```

```python
## categorize my data based on other categories

sns.boxplot(x='total_bill',y='day', hue='smoker', data=df, palette='bright')
```

### Violin Plot

Violin plot helps us to distribution od data in term of kernel density estimation and the box plot

```python
sns.violinplot(x='total_bill', y='day', data=df, palette='rainbow' )
```
